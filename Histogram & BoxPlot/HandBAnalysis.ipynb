```python
import pandas as pd 
import matplotlib.pyplot as plt 
from numpy import arange 


reviews = pd.read_csv("fandango_scores.csv")

cols = ["FILM", "RT_user_norm", "Metacritic_user_nom", "IMDB_norm", "Fandango_Ratingvalue", "Fandango_Stars"]
norm_reviews = reviews[cols]

print(norm_reviews[:5])
```

```python
#Building a freq Table

fandango_freq = norm_reviews["Fandango_Ratingvalue"].value_counts()
fandango_distribution = fandango_freq.sort_index()

imdb_freq = norm_reviews["IMDB_norm"].value_counts()
imdb_distribution = imdb_freq.sort_index()

print(fandango_distribution, imdb_distribution)

```

```python
#Generating basic histogram
fig, ax = plt.subplots()

ax.hist(norm_reviews['Fandango_Ratingvalue'], bins=20, range=(0,5))
plt.show()

```

```python
#Multiple Histogram bin practice

fig = plt.figure(figsize=(5,20))
ax1 = fig.add_subplot(4,1,1)
ax2 = fig.add_subplot(4,1,2)
ax3 = fig.add_subplot(4,1,3)
ax4 = fig.add_subplot(4,1,4)

ax1.hist(norm_reviews['Fandango_Ratingvalue'], bins=20, range=(0,5))
ax1.set_title("Distribution of Fandango Ratings")
ax1.set_ylim(0,50)

ax2.hist(norm_reviews['RT_user_norm'], bins=20, range=(0,5))
ax2.set_title("Distribution of Rotten Tomatoes Ratings")
ax2.set_ylim(0,50)

ax3.hist(norm_reviews['Metacritic_user_nom'], bins=20, range=(0,5))
ax3.set_title("Distribution of Metacritic Ratings")
ax3.set_ylim(0,50)

ax4.hist(norm_reviews['IMDB_norm'], bins=20, range=(0,5))
ax4.set_title("Distribution of IMDB Ratings")
ax4.set_ylim(0,50)

plt.show()
```

```python
#Boxplot and IQR BAsics

fig, ax = plt.subplots()

ax.boxplot(norm_reviews['RT_user_norm'])
ax.set_ylim(0,5)
ax.set_xticklabels(['Rotten Tomatoes']) #must be iterable object

plt.show()
```

```python
#Multiple box & whisker diagrams in same plot
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue']

fig, ax = plt.subplots()
ax.boxplot(norm_reviews[num_cols].values)
ax.set_xticklabels(num_cols, rotation=90)
ax.set_ylim(0,5)

plt.show()
```
