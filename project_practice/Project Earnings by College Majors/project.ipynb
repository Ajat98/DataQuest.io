```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.iloc[0]
recent_grads.describe
```

```python
recent_grads = recent_grads.dropna()
recent_grads.describe
#1 row removed as was missing values
```

```python
#Plotting various relations
#Sample size vs median
ax1 = recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title="Sample_size vs Median", figsize=(5,10) )
```

```python
#Sample size vs Unemployment
ax2 = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Sample_size vs Unemployment_rate', figsize=(5,10) )
```

```python
#Full Time vs median
ax3 = recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full time vs Median', figsize=(5,10) )
```

```python
#ShareWomen vs Unemployment
ax4 = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='ShareWomen vs Unemployment_rate', figsize=(5,10) )
```

```python
#Men and Median
ax5 = recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs Median', figsize=(5,10) )
```

```python
#Sample size vs median
ax6 = recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs Median', figsize=(5,10) )
```

```python
#Histograms
#need to label axes correctly, will do soon.

cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]
fig = plt.figure(figsize=(5,12))

for i in range(1,5):
    ax = fig.add_subplot(4,1, i)
    ax = recent_grads[cols[i]].plot(kind='hist', rot=40)

fig.tight_layout(pad=1.0)
```

```python
fig = plt.figure(figsize=(5,12))

for i in range(4,8):
    ax = fig.add_subplot(4,1,i-3)
    ax = recent_grads[cols[i]].plot(kind='hist', rot=40)

fig.tight_layout(pad=1.0)
```
